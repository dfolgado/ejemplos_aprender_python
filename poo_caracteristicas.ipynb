**Encapsulamiento**

```python
class Persona:

 def __init__(self, nombre):
   self.__nombre = nombre

 def get_nombre(self):
   return self.__nombre
```

**Herencia**

```python
class Animal:

 def __init__(self, nombre):
   self.__nombre = nombre

 def hacer_sonido(self):
   print("Sonido genérico")

class Perro(Animal):

 def hacer_sonido(self):
   print("Guau")


# Uso
p = Perro("Toby")
p.hacer_sonido()
```

**Herencia multinivel**

```python
class Vehiculo:
   pass


class Coche(Vehiculo):
   pass


class Deportivo(Coche):
   pass
```

**Polimorfismo**

```python
class Gato(Animal):
 def hacer_sonido(self):
   print("Miau")

toby = Perro("Toby")
toby.hacer_sonido()

cattana = Gato("Cattana")
cattana.hacer_sonido()
```

**super()**

```python
class Animal:
 def __init__(self, nombre):
   self.__nombre = nombre

class Perro(Animal):
  def __init__(self, nombre, raza):
    super().__init__(raza)
    self.raza = raza
```

**Métodos especiales (mágicos)**

```python
class Persona:
 def __init__(self, nombre):
   self.nombre = nombre

 def __str__(self):
   return f"Persona: {self.nombre}"

paco = Persona("Paco")
print(paco)
```


**Métodos estáticos**

```python
class Calculadora:
  @staticmethod
  def sumar(a, b):
    return a +  b

print(Calculadora.sumar(3,5))
```
